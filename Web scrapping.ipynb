```python
import re

def remove_emojis(string):
    emoj = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
    return re.sub(emoj, '', string)

old_reviews = []
old_reviews = list(temp['Reviews'].values)

new_reviews=[]

for i in range(len(old_reviews)):
    new_reviews.append(remove_emojis(old_reviews[i]))
    temp=[]
    
    for word in new_reviews[i]:
        if(word not in string.punctuation):
            temp.append(word)
    new_reviews[i] = ''.join(temp)
        
new_reviews
```

```python
df['Review Text'] = new_reviews
del df['Reviews']
df.rename(columns={'Review Text': 'Review'},inplace=True)

df['Scores'] = df['Review'].apply(lambda review: sid.polarity_scores(review))

df['compound']  = df['Scores'].apply(lambda score_dict: score_dict['compound'])
df['comp_score'] = df['compound'].apply(lambda c: 'positive' if c >=0 else 'negative')
df
```

```python
followers = []
positive_ratings = []
avg_selling_units = []
listings = []
avg_prices = []

def store_info():
    try:
        all_prices = []
        time.sleep(7)
        while(True):
            try:
                try:  #click profile button
                    driver.find_element(By.XPATH,'//*[@id="pi-component-container"]/div/div[2]/div/div/div/div[2]/div[2]/div/div/div[2]/div[3]').click()
                    break
                except:
                    driver.find_element(By.XPATH,'//*[@id="pi-component-container"]/div/div[2]/div/div/div/div[2]/div[2]/div/div/div[1]/div[3]').click()
                    break
            except:
                pass

        time.sleep(4)

        followers.append(driver.find_element(By.XPATH,'//*[@id="pi-component-container"]/div/div[2]/div/div/div/div[2]/div[1]/div/div/div/div[2]/div[2]/div').text)
        positive_ratings.append(driver.find_element(By.XPATH,'//*[@id="pi-component-container"]/div/div[2]/div/div/div/div[2]/div[1]/div/div/div/div[2]/div[3]').text)
        avg_selling_units.append(driver.find_element(By.CLASS_NAME,'based-on-label').text.split()[2])

        while(True):
            try:
                try:  #search for product inside the store
                    driver.find_element(By.XPATH,'//*[@id="pi-component-container"]/div/div[2]/div/div/div/div[2]/div[2]/div/div/div[3]/div/input').send_keys(search_term + Keys.RETURN)
                    break
                except:
                    driver.find_element(By.XPATH,'//*[@id="pi-component-container"]/div/div[2]/div/div/div/div[2]/div[2]/div/div/div[2]/div/input').send_keys(search_term + Keys.RETURN)
                    break
            except:
                pass

        time.sleep(4)
        listings.append(driver.find_element(By.CLASS_NAME,'tips--QRnmZ').text.split()[0])
        time.sleep(7)

        Temp = driver.find_elements(By.CLASS_NAME,'price--NVB62')
        for i in Temp:
            all_prices.append(i.text[3:].replace(',','')) #find all prices

        time.sleep(7)    
        all_prices = [int(i.strip()) for i in all_prices]
        avg_prices.append(np.mean(all_prices))  #find and append avg price
        time.sleep(8)

        driver.back()
        time.sleep(3)
        driver.back()
        time.sleep(1)
        driver.back()
        time.sleep(9)
    except: 
        print('Error')
        pass
```
